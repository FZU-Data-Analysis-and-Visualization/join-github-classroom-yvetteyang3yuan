### 一、基础创建与查看

1. 导入 pandas，用字典创建 DataFrame：

   name: Tom, Jerry, Mike

   score: 85, 92, 78

   gender: 男，女，男

```python
import pandas as pd
data={'name':['Tom','Jerry','Mike'],
    'score':[85,92,78],
    'gender':['男','女','男']} # 字典保存数据 字典的键变成DataFrame的列名
df=pd.DataFrame(data) # 表格数据 3x3
print(df)
```

2. 查看上述 DataFrame 的前 1 行、形状、列名、数据类型。

```python
print(df.head(1))
print()
print(df.shape)
print()
print(df.columns)
print()
print(df.dtypes)
```

3. 用列表创建 DataFrame：

   data = [[1, "A"], [2, "B"], [3, "C"]]

   列名设为 id、tag

```python
data=[[1,"A"],[2,"B"],[3,"C"]] # 列表创建
df2=pd.DataFrame(data,columns=['id','tag']) # 设置列名
print(df2)
```

### 二、列与行操作

1. 选取 score 和 gender 两列。

```python
result1=df[['score','gender']] 
print(result1)
```

2.选取索引 0、2 的行。

```python
result2=df.loc[[0,2]]
print(result2)
```

3.筛选 score>80 的行。

```python
result3=df[df['score']>80]
print(result3)
```

4.新增一列 grade：score≥90 为优秀，否则为良好。

```python
df['grade']=df['score'].apply(lambda x:'优秀' if x>=90 else '良好')
print(df)
```

5.删除 gender 列。

```python
df_dele=df.drop(columns=['gender'])
print(df_dele)
```

### 五、排序与去重

1. 按 score 升序排序。

```python
df_sort = df.sort_values(by='score')
print(df_sort)
```

2. 给 DataFrame 添加一行重复数据（如 Tom,85, 男），去重。

```python
new_row = pd.DataFrame({
    'name': ['Tom'],
    'score': [85],
    'gender': ['男'],
    'grade':['良好']
})
df_repeat = pd.concat([df, new_row], ignore_index=True)
print(df_repeat)
```

### 六、合并与拼接

1. 横向合并：

   df1 = pd.DataFrame ({"id":[1,2], "name":["X","Y"]})

   df2 = pd.DataFrame ({"id":[1,2], "age":[20,30]})

   按 id 合并。

```python
import pandas as pd
df1 = pd.DataFrame({
    "id": [1, 2],
    "name": ["X", "Y"]
})
df2 = pd.DataFrame({
    "id": [1, 2],
    "age": [20, 30]
})
# 按 id 合并
df_merge = pd.merge(df1, df2, on="id")
print(df_merge)
```


2. 纵向拼接：

   df_a = pd.DataFrame ({"val":[1,2]})

   df_b = pd.DataFrame ({"val":[3,4]})

   拼接后重置索引。

```python
df_a = pd.DataFrame({
    "val": [1, 2]
})
df_b = pd.DataFrame({
    "val": [3, 4]
})
# 纵向拼接并重置索引
df_concat = pd.concat([df_a, df_b], ignore_index=True)
print(df_concat)
```

### 七、时间处理

1. 创建时间列：["2025-01-01","2025-01-02","2025-01-03"]，转为 datetime 格式。


```python
import pandas as pd
df = pd.DataFrame({
    "date": ["2025-01-01", "2025-01-02", "2025-01-03"]
})
df["date"] = pd.to_datetime(df["date"])
print(df)
print(df.dtypes)
```


2. 提取时间列的月份、星期。

```python
df["month"] = df["date"].dt.month
df["weekday"] = df["date"].dt.weekday
print(df)
```

### 八、高级操作

1. 对 score 列做标准化（(值 - 均值)/ 标准差）。


```python
import pandas as pd
data = {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df = pd.DataFrame(data)
print(df)
df['score_standard'] = (df['score'] - df['score'].mean()) / df['score'].std()
print(df)
```
